# Suche mit regulären Ausdrücken

```python
import codecs

buch = codecs.open('BrückenkursPhysik.txt','r','utf8')
for zeile in buch:
    if "kraft" in zeile:
        print(zeile)
buch.close()
```

```python
import re

buch = codecs.open('BrückenkursPhysik.txt','r','utf8')
for zeile in buch:
    if re.search(r'\b\S*[Kk]r(a|ä)fte?\b',zeile):
        print(zeile)
buch.close()
```

## Was wurde gefunden?

```python
text = """In einer Gesellschaft, die durch Zentralregierungen
und Globalisierung geprägt ist, können diese Sprachen sich 
nur schwer behaupten und jährlich sterben schätzungsweise
einige dutzende Sprachen aus."""

fund = re.search(r'ge\S+t',text)

print(fund.group(0))
```

Das Gleiche kombiniert mit dem ersten Beispiel

```python
buch = codecs.open('BrückenkursPhysik.txt','r','utf8')
for zeile in buch:
    fund = re.search(r'\b\S*[Kk]r(a|ä)fte?\b',zeile)
    if  fund:
        print(fund.group(0))
buch.close()
```

Die Liste ist etwas lang. Spannender wäre es jetzt zusammenzufassen uns zu zählen. In Python geht das ganz leicht:

```python
from collections import Counter
import codecs
import re

kraftliste = []

buch = codecs.open('BrückenkursPhysik.txt','r','utf8')
for zeile in buch:
    fund = re.search(r'\b\S*[Kk]r(a|ä)fte?\b',zeile)
    if  fund:
        kraftliste.append(fund.group(0))
buch.close()
kraftstatistik = Counter(kraftliste)
kraftstatistik.most_common() 
```

## Wo wurde gefunden?

```python
fund = re.search(r'ge\S+t',text)

print(fund.start() , "--"  , fund.end())
```

## Gruppen

```python
fund =  re.search(r'ge(\S+)t',text)
verb = fund.group(1) + "en"

print(verb, "(Stamm gefunden auf:",fund.start(1), "--",fund.end(1),")")
```

## Findall

```python
fundliste =  re.findall(r'[Gg]e\S+t',text)
for fund in fundliste:
    print(fund)
```

```python
fundliste =  re.findall(r'([Gg]e)(\S+)(t)',text)
for fund in fundliste:
    print(fund[0],"-",fund[1],"-",fund[2])
```

Ein kleiner Trick:

Wenn wir das ganze gefundene Wort haben wolle, fügen wir einfach ein weitere Klammerpaar zu!

```python
fundliste =  re.findall(r'(([Gg]e)(\S+)(t))',text)
for fund in fundliste:
    print(fund[0])
```

## Split

```python
tokens =  re.split(r'[,. \r\n]+',text)

print(tokens)
```

# Übungen

Mit dem folgende Code suchen wir Akronyme von 3 oder 4 Buchstaben und zeigen diese im Kontext. Dieses Anzeigeformat nennt man KWIC: Keyword in Context.

```python
datei = codecs.open('BrückenkursPhysik.txt','r','utf8')

nr = 0
for zeile in datei:
    nr = nr+1
    fundstelle = re.search(r'\b[A-Z]{3,4}\b',zeile)
    if fundstelle:
        start = fundstelle.start()
        end = fundstelle.end()
        linker_kontext = zeile[max(0,start-20):start]
        linker_kontext = (20 - len(linker_kontext))*' ' + linker_kontext
        rechter_kontext = zeile[end:end+20]
        print("Zeile", nr,linker_kontext,fundstelle.group(0),rechter_kontext,sep='\t')
    
datei.close()
```
